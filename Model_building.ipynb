```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN
```

```python
df = pd.read_csv('tel_churn.csv')
```

```python
df.head(20)
```

```python
df = df.drop("Unnamed: 0", axis = 1)
```

```python
df.head(20)
```

```python
#create x and y variables

x = df.drop('Churn',axis = 1)
x
```

```python
#creating X & Y variables

y = df['Churn']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

*`Decition tree classifier`*

```python
model_dt = DecisionTreeClassifier(criterion = 'gini', random_state = 100, max_depth = 6, min_samples_leaf = 8)
```

```python
model_dt.fit(x_train,y_train)
```

```python
y_predict = model_dt.predict(x_test)
```

```python
y_predict
```

```python
model_dt.score(x_test,y_predict)
```

```python
print(classification_report(y_test,y_predict,labels = [0,1]))
```

```python
sm = SMOTEENN()
x_resampled, y_resampled = sm.fit_resample(x, y)
```

```python
xk_train, xk_test, yk_train, yk_test = train_test_split(x_resampled,y_resampled,test_size = 0.2)
```

```python
model_dt_smote = DecisionTreeClassifier(criterion = 'gini', random_state = 100, max_depth = 6, min_samples_leaf = 8)
```

```python
model_dt_smote.fit(xk_train, yk_train)
```

```python
y_predict_smote = model_dt_smote.predict(xk_test)
```

```python
print(classification_report(yk_test,y_predict_smote, labels = [0,1])) 
```

```python
print(confusion_matrix(yk_test,y_predict_smote)) 
```
